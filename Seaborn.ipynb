# Seaborn:
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics. Seaborn is particularly useful for creating informative and visually appealing statistical graphics. It builds on top of matplotlib and integrates closely with pandas data structures.

Seaborn simplifies the process of creating complex visualizations such as scatter plots, bar plots, box plots, histograms, heatmaps, and more. It also offers built-in themes and color palettes to enhance the aesthetics of the plots. Seaborn is often preferred for its ease of use and its ability to generate visually appealing plots with relatively little code.

## plots:

Seaborn implements two types of functions, figure-level and axes-level, to provide users with flexibility, convenience, and control over the plotting process.

**Figure-Level Function:**

- Figure-level functions in Seaborn, such as `sns.catplot()` and `sns.relplot()`, create entire figures or grids of plots.
- These functions typically involve creating a new figure or grid and then plotting multiple axes or subplots within that figure or grid.
- They offer high-level control over the overall structure and layout of the plots, such as arranging multiple subplots in a grid or creating facets based on categorical variables.
- Figure-level functions are useful for creating complex visualizations involving multiple subplots or facets, as they provide convenient ways to manage the overall layout and structure of the plots.

**Axes-Level Function:**

- Axes-level functions in Seaborn, such as `sns.scatterplot()` and `sns.barplot()`, create individual plots on a specific set of axes within an existing figure.
- These functions typically involve plotting data on an existing set of axes, allowing for fine-grained control over the appearance and customization of the plot.
- They offer low-level control over the properties of the individual plot elements, such as colors, markers, and axis labels.
- Axes-level functions are useful for customizing the appearance and behavior of individual plots, as they allow for direct manipulation of the axes and plot elements.

In simple terms, figure-level functions are used to create entire figures or grids of plots, while axes-level functions are used to create individual plots on specific sets of axes within those figures or grids. Figure-level functions provide high-level control over the overall layout and structure of the plots, while axes-level functions provide low-level control over the appearance and customization of individual plots.


<b>Relative merits of figure-level functions</b></br>

Advantages:</br>
* Easy faceting by data variables</br>

* Easy figure-level customization</br>

## 1.a) Scatter Plot:

A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It is composed of individual data points, where each point represents the value of both variables. The position of each point on the plot corresponds to the values of the two variables, with one variable plotted along the x-axis and the other along the y-axis.

**Definition:**
A scatter plot is a graphical representation of the relationship between two continuous variables. It is commonly used to visualize patterns, trends, and correlations between variables.

**When to use:**
- **Exploring Relationships:** Scatter plots are useful for exploring the relationship between two variables, such as determining if there is a linear or nonlinear association between them.
- **Identifying Patterns:** They help in identifying patterns, clusters, or outliers within the data.
- **Assessing Correlation:** Scatter plots are often used to assess the strength and direction of the correlation between variables.
- **Visualizing Distribution:** They provide a visual representation of the distribution of data points across the range of both variables.

## 2.b) KDE plot:

A KDE (Kernel Density Estimation) plot is a type of data visualization that estimates the probability density function of a continuous random variable. It represents the distribution of the data by smoothing out the histogram, providing a smooth curve that approximates the underlying probability density.

**Definition:**
A KDE plot is a graphical representation of the estimated probability density function of a continuous random variable. It is created by smoothing out the histogram of the data, typically using a kernel function, to provide a continuous curve that approximates the distribution of the data.

**When to use:**
- **Visualizing Distribution:** KDE plots are useful for visualizing the distribution of data, especially when the underlying distribution is not well-known or when the data is noisy.
- **Comparing Distributions:** They allow for the comparison of multiple distributions on the same plot, making it easy to compare the shapes and spread of different datasets.
- **Assessing Smoothness:** KDE plots provide a smooth representation of the data distribution, which can help in assessing the smoothness of the underlying distribution and identifying potential patterns or features.
- **Identifying Peaks and Valleys:** They help in identifying peaks, valleys, and modes in the data distribution, providing insights into the central tendency and variability of the data.


```python
# Figure level function
sns.displot(data=df, x="etest_p", kind="kde", hue="gender", fill=True)
plt.show()

# Axes-level function
# fig, ax = plt.subplots()
# sns.kdeplot(data=df, x="etest_p", hue="gender", fill=True, ax=ax)
# plt.show()
```

## 2.c) Rug plot:

A rug plot is a type of data visualization that displays the distribution of data points along a single axis. It is typically used in conjunction with other plots, such as histograms or KDE plots, to provide additional information about the density and distribution of data points.

**Definition:**
A rug plot is a graphical representation of the distribution of data points along a single axis. It consists of a series of marks, or "rugs," placed at the location of each data point along the axis. Each rug mark represents the presence of a data point at that location.

**When to use:**
- **Enhancing Other Plots:** Rug plots are often used in conjunction with other plots, such as histograms or KDE plots, to provide additional information about the density and distribution of data points.
- **Visualizing Individual Data Points:** They help in visualizing individual data points and their distribution along the axis, providing insights into the density and spread of the data.
- **Assessing Data Density:** Rug plots provide a visual representation of the density of data points at different locations along the axis, making it easy to identify regions of high or low density.
- **Comparing Distributions:** They can be used to compare the distribution of data points across different groups or categories, especially when overlaid on other plots.

## 4.bi) Boxplot:

A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a continuous variable through its quartiles. It provides a visual summary of the central tendency, variability, and skewness of the data distribution, as well as the presence of outliers.

**Definition:**
A box plot is a graphical representation of the distribution of a continuous variable. It consists of a rectangular box, known as the interquartile range (IQR), which spans the first quartile (Q1) to the third quartile (Q3) of the data. A line within the box represents the median (Q2) of the data. Additionally, "whiskers" extend from the box to the minimum and maximum values within a specified range, typically 1.5 times the IQR. Outliers beyond this range are plotted individually as points.

**When to use:**
- **Visualizing Distribution:** Box plots are useful for visualizing the distribution of a continuous variable, providing insights into its central tendency, spread, and skewness.
- **Comparing Groups:** They allow for the comparison of the distribution of a variable across different groups or categories, making it easy to identify differences or similarities.
- **Identifying Outliers:** Box plots help in identifying outliers or extreme values in the dataset, which may indicate unusual behavior or measurement errors.
- **Assessing Symmetry:** They provide a visual indication of the symmetry or skewness of the data distribution, based on the position of the median relative to the quartiles.


```python
# Figure level function
sns.catplot(data=df, y="etest_p", kind="box")
plt.show()

# Axes-level function
# fig, ax = plt.subplots()
# sns.boxplot(data=df, y="etest_p", ax=ax)
# plt.show()
```

## 5.b) Residplot:

A residual plot is a type of data visualization used in regression analysis to assess the goodness-of-fit of a regression model. It displays the residuals, or the differences between observed and predicted values, against the independent variable(s) or the fitted values.

**Definition:**
A residual plot is a graphical representation of the residuals, which are the differences between the observed values and the predicted values obtained from a regression model. The residuals are plotted against the independent variable(s) or the fitted values from the model. Residual plots are useful for diagnosing potential issues with the regression model, such as heteroscedasticity, nonlinearity, or outliers.

**When to use:**
- **Assessing Model Assumptions:** Residual plots are used to assess whether the assumptions of the regression model, such as linearity and constant variance of residuals, are met.
- **Detecting Patterns:** They help in detecting patterns or systematic errors in the residuals, such as nonlinearity or heteroscedasticity, which may indicate that the model is misspecified.
- **Identifying Outliers:** Residual plots can reveal the presence of outliers or influential data points that may have a disproportionate impact on the model.
- **Validating Model Fit:** They are used for validating the fit of the regression model to the data and identifying areas where the model may need improvement.

## 6.a) Facet Grid:

A facet grid is a type of data visualization that allows for the creation of multiple plots or subplots arranged in a grid layout, with each plot representing a subset of the data based on one or more categorical variables.

**Definition:**
A facet grid is a graphical representation of data where multiple plots or subplots are arranged in a grid layout, typically based on one or more categorical variables. Each subplot represents a subset of the data corresponding to a unique combination of categorical values, allowing for comparisons between different groups or categories within the dataset.

**When to use:**
- **Comparing Categorical Variables:** Facet grids are useful for comparing the distribution or relationship between variables across different categories or groups within the dataset.
- **Visualizing Multivariate Relationships:** They help in visualizing multivariate relationships by arranging plots based on combinations of categorical variables.
- **Handling Large Datasets:** Facet grids are effective for visualizing large datasets by breaking them down into smaller, more manageable subsets.
- **Identifying Patterns:** They can reveal patterns or trends within the data by providing a structured way to explore relationships between variables.


```python
object = sns.FacetGrid(data=df,row="gender",col="status")
object.map(sns.scatterplot,"hsc_p","ssc_p")
plt.show()
```

## 6.b) Pair Plot:

A pair plot is a type of data visualization that displays pairwise relationships between multiple variables in a dataset. It consists of a grid of scatter plots, where each plot represents the relationship between two variables, and histograms or kernel density plots along the diagonal, which display the distribution of each variable.

**Definition:**
A pair plot is a graphical representation of pairwise relationships between multiple variables in a dataset. It displays scatter plots of each pair of variables against each other in a grid layout. Along the diagonal of the grid, histograms or kernel density plots are typically shown to represent the distribution of each variable.

**When to use:**
- **Exploring Relationships:** Pair plots are useful for exploring relationships between multiple variables in a dataset, providing insights into correlations, trends, and patterns.
- **Identifying Outliers:** They help in identifying outliers or unusual observations by visualizing the distribution of data points across different combinations of variables.
- **Comparing Variables:** Pair plots allow for the comparison of multiple variables simultaneously, making it easy to identify similarities or differences between them.
- **Assessing Multicollinearity:** They can be used to assess multicollinearity by examining the strength and direction of relationships between variables.


```python
sns.pairplot(df)
plt.show()
```

## 6.c) PairGrid:

A pair grid is a type of data visualization that extends the concept of a pair plot by allowing for more customization and flexibility in the arrangement of pairwise plots. It enables the creation of a grid of subplots, with each subplot representing the relationship between two variables, and allows for the inclusion of additional plot types and annotations.

**Definition:**
A pair grid is a graphical representation of pairwise relationships between multiple variables in a dataset, similar to a pair plot. However, it offers more customization options by allowing users to specify the arrangement of subplots, the types of plots to be displayed, and additional annotations or overlays.

**When to use:**
- **Customization:** Pair grids are useful when customization of the pairwise plots is required, such as specifying the order of variables, customizing plot aesthetics, or including additional annotations.
- **Complex Relationships:** They are effective for visualizing complex relationships between variables by arranging plots in a grid layout and allowing for the inclusion of different plot types, such as scatter plots, line plots, or kernel density plots.
- **Comparing Groups:** Pair grids enable the comparison of pairwise relationships between variables across different groups or categories within the dataset by facetting the grid based on categorical variables.
- **Visualizing Multivariate Data:** They help in visualizing multivariate data by providing a structured way to explore relationships between multiple variables and identify patterns or trends.


```python
obj = sns.PairGrid(data=df)
obj.map_diag(sns.histplot)
obj.map_upper(sns.scatterplot)
obj.map_lower(sns.kdeplot)
plt.show()
```

```python
var = ["etest_p","mba_p"]
obj2 = sns.PairGrid(data=df[var])
obj2.map_diag(sns.kdeplot)
obj2.map_offdiag(sns.scatterplot)
plt.show()
```

## 7.a) Joint plot:

A joint plot is a type of data visualization that displays the relationship between two variables along with their marginal distributions. It combines multiple plots, such as scatter plots, histograms, or kernel density plots, into a single visualization to provide insights into the joint and marginal distributions of the variables.

**Definition:**
A joint plot is a graphical representation of the relationship between two variables, typically using a scatter plot to show their joint distribution. In addition to the scatter plot, it also includes marginal plots along the axes to display the distribution of each variable individually, such as histograms or kernel density plots.

**When to use:**
- **Visualizing Relationships:** Joint plots are useful for visualizing the relationship between two variables and identifying patterns or trends in their joint distribution.
- **Assessing Correlation:** They help in assessing the strength and direction of the correlation between variables by displaying the scatter plot of their joint distribution.
- **Examining Marginal Distributions:** Joint plots provide insights into the marginal distributions of the variables by including histograms or kernel density plots along the axes.
- **Detecting Outliers:** They can reveal the presence of outliers or unusual observations in the data by examining the scatter plot of the joint distribution.


```python
sns.jointplot(data=df,x="hsc_p",y="ssc_p")
plt.show()
```
